## 1. Calculate ELO ratings for all teams

```python
results = pd.read_csv('../data/raw/international_results.csv')
results['date'] = pd.to_datetime(results['date'])

# Filter to post-1990 for more relevant ratings
results = results[results['date'].dt.year >= 1990]

elo_df = calculate_elo_ratings(results)
print(f'ELO ratings for {len(elo_df)} teams')
elo_df.head(20)
```

```python
# Plot top 20 teams by ELO
top20 = elo_df.head(20)
plt.figure(figsize=(12, 6))
plt.barh(top20['team'][::-1], top20['elo'][::-1], color='steelblue', edgecolor='navy')
plt.xlabel('ELO rating')
plt.title('Top 20 international teams by ELO rating')
plt.tight_layout()
plt.show()
```
