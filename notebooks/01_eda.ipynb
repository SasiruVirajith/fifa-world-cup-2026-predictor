```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (12, 5)
print('Libraries loaded ✓')
```

## 1. Load StatsBomb match data

```python
# Load WC 2022 matches
try:
    matches = pd.read_csv('../data/raw/statsbomb_matches_wc2022.csv')
    print(f'WC 2022: {len(matches)} matches')
    matches.head()
except FileNotFoundError:
    print('Run src/data_pipeline.py first to fetch data')
```

```python
# What columns do we have?
print(matches.columns.tolist())
matches.dtypes
```

## 2. Goals distribution — how many goals per match?

```python
if 'home_score' in matches.columns:
    matches['total_goals'] = matches['home_score'] + matches['away_score']
    matches['total_goals'].hist(bins=10, edgecolor='black', color='steelblue')
    plt.title('Goals per match — WC 2022')
    plt.xlabel('Total goals')
    plt.ylabel('Number of matches')
    plt.show()
    print(f'Average goals per match: {matches["total_goals"].mean():.2f}')
```
